# Deciphering and Analysis of Shakespeare play

## Overview

In this project, I am tasked with deciphering a given Shakespeare text, I begin with cleaning the translated text data by taking care of unwanted characters and formatting. After this, I create visualisations that highlight key features of the text, such as word length, words per line and most frequent words. It is useful to note that due to it being for a first year module, the only libraries I was permitted to use were `matplotlib` and `numpy`.

## The cipher

The shift of the cipher was determined by the greatest common divisor of the `user_number` which was pre-set to be 506 and the value $k$ such that
$$
    \sum_{n=1}^{k}\frac{1}{\sqrt[5]{n}} > 506 \,.
$$

To do this I first wrote a function containing the sum above, then used the result within the `find_shift` function to find the value of $k$. In the same function I also wrote Euclid's algorithm to compute the GCD of the value of $k$ and `user_number` which is the shift.
